```python
import pandas as pd
import json
import requests
## Functions
#Get ndb product number | USDA API 
def ndb_search(product):
    url = 'https://api.nal.usda.gov/ndb/search'
    key = 'uC7EeReDRVWw7Kzijvvn8LtCDWR3D6hPyksxuFTW'
    params = (('format', 'json'),('q', product),('api_key', key))
    response = requests.get(url, params = params)
    products = response.json()['list']['item']
    products_df = pd.DataFrame(products)     
    product_ndb = pd.DataFrame(products_df['ndbno']).head(1)
    ndb_int = int(product_ndb['ndbno'])
    
    return ndb_int

#Get product ingredients list using NDB number | USDA API 
def ndb_report(product_ndb):
    url = 'https://api.nal.usda.gov/ndb/V2/reports'
    key = 'uC7EeReDRVWw7Kzijvvn8LtCDWR3D6hPyksxuFTW'
    params = (('ndbno', product_ndb),('type', 'f'),('format', 'json'),('api_key', key))
    url = 'https://api.nal.usda.gov/ndb/V2/reports'
    response = requests.get(url, params = params)
    data_list = response.json()['foods'][0]

    ingredients = data_list['food']['ing']
  
    return ingredients['desc'].split(',')
    return ingredients

```
