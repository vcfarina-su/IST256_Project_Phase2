```python
def CP_data():   
    html = requests.get('https://toxnet.nlm.nih.gov/cpdb/chemnameindex.html')
    soup = BeautifulSoup(html.text,'html.parser')
    text = soup.get_text(strip = True)
    return text

CP_data = CP_data()[1831:59244]
CP_list = CP_data.split()
#print(CP_data)

CP_chems = []

for i in CP_list:
    if i != '(CAS':
        CP_chems.append(i)

CP_chems

```

```python
CP_nums = []
for i in CP_chems:
    if i != i.isalpha():
        CP_nums.append(i)

CP_nums2 = []
leading_char = []
CP_text = []
for i in CP_nums:
    remove_parenth = i.replace(')',' ')
    CP_nums2.append(remove_parenth)
    
for i in CP_nums2:
    space = i.find(' ')
    leading_char = i[space: ]
    CP_text.append(leading_char)
CP_text
CP_df = pd.DataFrame(CP_text)

#CP_df

CPnum_df = pd.DataFrame(CP_chems)

CPnum_df
```
